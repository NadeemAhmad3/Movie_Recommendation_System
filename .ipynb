```python


def create_numerical_analysis_plots(data, numerical_cols):
    """
    Creates a single, comprehensive figure analyzing a list of numerical columns.
    For each column, it shows a distribution plot (histogram + KDE) and a box plot side-by-side.
    """
    n_cols = len(numerical_cols)
    fig, axes = plt.subplots(n_cols, 2, figsize=(16, 6 * n_cols), gridspec_kw={'width_ratios': [3, 1]})
    fig.suptitle('Numerical Variables: Distribution and Outlier Analysis', fontsize=18, fontweight='bold')

    for i, col in enumerate(numerical_cols):
        ax_hist, ax_box = axes[i, 0], axes[i, 1]

        # --- Distribution Plot (Histogram + KDE) ---
        sns.histplot(data[col], kde=True, ax=ax_hist, color='skyblue', bins=40)
        ax_hist.set_title(f'{col.title()} Distribution', fontsize=14)
        ax_hist.set_xlabel('')
        ax_hist.set_ylabel('Frequency')
        mean_val = data[col].mean()
        median_val = data[col].median()
        ax_hist.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
        ax_hist.axvline(median_val, color='green', linestyle='-', label=f'Median: {median_val:.2f}')
        ax_hist.legend()

        # --- Box Plot ---
        sns.boxplot(y=data[col], ax=ax_box, color='lightcoral')
        ax_box.set_title(f'{col.title()} Box Plot', fontsize=14)
        ax_box.set_xlabel('')
        ax_box.set_ylabel('')

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    return fig

def display_summary_statistics(data, numerical_cols):
    """
    Calculates and displays a styled summary DataFrame for numerical columns.
    """
    summary_df = data[numerical_cols].describe().T
    summary_df['skewness'] = data[numerical_cols].skew()
    summary_df['kurtosis'] = data[numerical_cols].kurt()
    
    # Apply styling for better readability
    return summary_df.style.background_gradient(cmap='viridis', low=0.5, high=0).format("{:,.2f}")

def display_correlation_analysis(data, numerical_cols, threshold=0.7):
    """
    Generates a correlation heatmap and efficiently finds and prints highly correlated pairs.
    """
    print("--- Correlation Matrix Heatmap ---")
    corr_matrix = data[numerical_cols].corr()
    
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, ax=ax)
    ax.set_title('Correlation Matrix of Numerical Variables', fontsize=16, fontweight='bold')
    plt.show()

    # --- Efficiently find highly correlated pairs ---
    print(f"\n--- Highly Correlated Pairs (Threshold > {threshold}) ---")
    # Unstack the correlation matrix, remove self-correlations, and filter
    sol = corr_matrix.abs().unstack()
    sol = sol[sol > threshold]
    sol = sol[sol < 1.0] # Remove self-correlation (corr=1)
    
    # The result contains duplicates (A-B, B-A), so we divide by 2
    if sol.empty:
        print("No highly correlated pairs found.")
    else:
        # Create a clean DataFrame from the results
        high_corr_df = pd.DataFrame(sol).reset_index()
        high_corr_df.columns = ['Variable 1', 'Variable 2', 'Correlation']
        # Remove duplicate pairs (e.g., keep ('age', 'rating') but drop ('rating', 'age'))
        high_corr_df['sorted_vars'] = high_corr_df.apply(lambda row: tuple(sorted((row['Variable 1'], row['Variable 2']))), axis=1)
        high_corr_df = high_corr_df.drop_duplicates(subset='sorted_vars').drop(columns='sorted_vars')
        display(high_corr_df)
        
print("Helper functions for Numerical EDA have been created.")
# Define the columns you wish to analyze from your original code
numerical_columns_to_analyze = ['rating', 'timestamp', 'age']

# --- 1. Generate and Display Distribution/Outlier Plots ---
numerical_plots_fig = create_numerical_analysis_plots(data, numerical_columns_to_analyze)
plt.show()

# --- 2. Display Styled Summary Statistics ---
print("--- Numerical Variables Summary Statistics ---")
summary_table = display_summary_statistics(data, numerical_columns_to_analyze)
display(summary_table)

# --- 3. Display Correlation Analysis ---
display_correlation_analysis(data, numerical_columns_to_analyze, threshold=0.7)
```

```python
def predict_user_cf_rating(user_id, movie_id, user_similarity_df, train_matrix, user_means, k=50):
    """
    Predicts a rating for a user-movie pair using an improved user-based CF algorithm.
    This function correctly uses the mean-centered similarity matrix.
    """
    # Fallback to the user's average rating or a global average of 3.0 if user/movie is unknown
    if user_id not in user_similarity_df.index or movie_id not in train_matrix.columns:
        return user_means.get(user_id, 3.0)

    # --- EFFICIENT FILTERING ---
    # 1. Get top k similar users (excluding the user themselves)
    similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:]

    # 2. Get the ratings for the target movie from all users
    movie_ratings = train_matrix[movie_id]

    # 3. Find similar users who have actually rated the movie
    top_similar_raters = similar_users.intersection(movie_ratings[movie_ratings > 0].index)[:k]

    # If no similar users have rated this movie, fall back to the user's average rating
    if top_similar_raters.empty:
        return user_means.get(user_id, 3.0)

    # --- CORRECT PREDICTION LOGIC for Mean-Centered Data ---
    # Get the similarity scores and the ratings of the top similar raters
    top_similarities = user_similarity_df.loc[user_id, top_similar_raters]
    top_ratings = train_matrix.loc[top_similar_raters, movie_id]
    
    # The predicted rating is the user's average rating plus a weighted average of
    # how much the similar users deviated from *their* average.
    # We use the raw (non-centered) ratings here for the weighted average calculation.
    numerator = np.dot(top_ratings, top_similarities)
    denominator = top_similarities.sum()

    if denominator == 0:
        return user_means.get(user_id, 3.0) # Fallback

    predicted_rating = numerator / denominator

    # Clip the result to be within the valid rating range [1, 5]
    return np.clip(predicted_rating, 1, 5)


def get_user_cf_recommendations(user_id, user_similarity_df, train_matrix, movies_df, user_means, n=10, k_similar_users=100):
    """
    Generates top N movie recommendations for a user with high efficiency.
    Instead of checking all unseen movies, it only checks movies rated by the most similar users.
    """
    if user_id not in train_matrix.index:
        print(f"User {user_id} not found in the dataset.")
        return pd.DataFrame() # Return empty DataFrame

    # --- EFFICIENT CANDIDATE SELECTION ---
    # 1. Get the top K most similar users
    top_k_similar_users = user_similarity_df[user_id].sort_values(ascending=False).index[1:k_similar_users+1]

    # 2. Get all movies rated by these similar users
    candidate_movies_ratings = train_matrix.loc[top_k_similar_users]

    # 3. Filter for movies that were rated by at least a few of them to find relevant items
    candidate_movies_ratings = candidate_movies_ratings.loc[:, (candidate_movies_ratings > 0).sum() > 2]

    # 4. Get the user's own rated movies to exclude them from recommendations
    user_rated_movies = train_matrix.loc[user_id][train_matrix.loc[user_id] > 0].index
    
    # 5. Determine the final list of candidate movies by removing ones the user has already seen
    candidate_movies = [movie for movie in candidate_movies_ratings.columns if movie not in user_rated_movies]

    # --- Predict ratings ONLY for the small, relevant candidate list ---
    predictions = []
    for movie_id in candidate_movies:
        pred_rating = predict_user_cf_rating(user_id, movie_id, user_similarity_df, train_matrix, user_means)
        predictions.append((movie_id, pred_rating))

    # Sort predictions and get top N
    predictions.sort(key=lambda x: x[1], reverse=True)
    top_n_recs = predictions[:n]
    
    # --- HELPFUL OUTPUT ---
    # Return a clean DataFrame with movie titles and predicted ratings
    if not top_n_recs:
        return pd.DataFrame(columns=['movie_id', 'predicted_rating', 'title'])

    rec_ids, rec_scores = zip(*top_n_recs)
    rec_df = pd.DataFrame({'movie_id': rec_ids, 'predicted_rating': rec_scores})
    
    # Merge with the movies DataFrame to get titles
    rec_df = rec_df.merge(movies_df, on='movie_id', how='left')
    
    return rec_df[['movie_id', 'title', 'predicted_rating']]
```

```python
def predict_item_cf_rating(user_id, movie_id, item_similarity_df, train_matrix, k=50):
    """
    Predicts a rating for a user-movie pair using an improved item-based CF algorithm.
    """
    # Fallback to a global average of 3.0 if user/movie is unknown
    if user_id not in train_matrix.index or movie_id not in item_similarity_df.index:
        return 3.0

    # --- EFFICIENT FILTERING ---
    # 1. Get the items the user has already rated
    user_ratings = train_matrix.loc[user_id]
    rated_items_by_user = user_ratings[user_ratings > 0].index

    # 2. Get the similarities between the target movie and the items the user has already rated
    similarities_to_rated_items = item_similarity_df.loc[movie_id, rated_items_by_user]
    
    # 3. Get the top K most similar items from that list
    top_k_similar_items = similarities_to_rated_items.nlargest(k)

    # If the user hasn't rated any items similar to the target movie, fall back
    if top_k_similar_items.empty:
        return 3.0

    # --- PREDICTION LOGIC ---
    # Get the user's ratings for these top similar items
    ratings_for_top_items = user_ratings.loc[top_k_similar_items.index]

    # Calculate the weighted average: sum(rating * similarity) / sum(similarity)
    numerator = np.dot(ratings_for_top_items, top_k_similar_items)
    denominator = top_k_similar_items.sum()

    if denominator == 0:
        return 3.0 # Fallback

    predicted_rating = numerator / denominator
    return np.clip(predicted_rating, 1, 5)


def get_item_cf_recommendations(user_id, train_matrix, item_similarity_df, movies_df, n=10):
    """
    Generates top N movie recommendations for a user with high efficiency.
    Instead of checking all unseen movies, it generates candidates from items similar to what the user likes.
    """
    if user_id not in train_matrix.index:
        print(f"User {user_id} not found.")
        return pd.DataFrame() # Return empty DataFrame

    # --- EFFICIENT CANDIDATE SELECTION ---
    # 1. Get items the user rated highly (e.g., rating >= 4)
    user_ratings = train_matrix.loc[user_id]
    highly_rated_items = user_ratings[user_ratings >= 4].index

    # 2. For each highly-rated item, find its most similar items to build a candidate pool
    candidate_items = set()
    for item in highly_rated_items:
        # Get top 15 similar items for each liked item
        similar_items = item_similarity_df[item].nlargest(15).index
        candidate_items.update(similar_items)

    # 3. Remove items the user has already seen
    rated_items = user_ratings[user_ratings > 0].index
    candidate_items.difference_update(rated_items)
    
    # --- Predict ratings ONLY for the small, relevant candidate list ---
    if not candidate_items:
        return pd.DataFrame(columns=['movie_id', 'title', 'predicted_rating'])
        
    predictions = []
    for movie_id in candidate_items:
        pred = predict_item_cf_rating(user_id, movie_id, item_similarity_df, train_matrix)
        predictions.append((movie_id, pred))

    # Sort predictions and get top N
    predictions.sort(key=lambda x: x[1], reverse=True)
    top_n_recs = predictions[:n]
    
    # --- HELPFUL OUTPUT ---
    if not top_n_recs:
        return pd.DataFrame(columns=['movie_id', 'title', 'predicted_rating'])

    rec_ids, rec_scores = zip(*top_n_recs)
    rec_df = pd.DataFrame({'movie_id': rec_ids, 'predicted_rating': rec_scores})
    
    # Perform a single, efficient merge to get movie titles
    rec_df = rec_df.merge(movies_df, on='movie_id', how='left')
    
    return rec_df[['movie_id', 'title', 'predicted_rating']]
```

```python
from sklearn.decomposition import TruncatedSVD

# --- 1. Train the SVD Model ---
# We reuse the `train_matrix_centered` we already created earlier. No need to re-calculate.
print("--- Training SVD Model ---")
n_components = 50
svd = TruncatedSVD(n_components=n_components, random_state=42)

# Fit SVD on the centered training data
user_factors = svd.fit_transform(train_matrix_centered) # Shape: (n_users, n_components)
item_factors = svd.components_ # Shape: (n_components, n_items)

print(f"SVD training complete. Explained variance with {n_components} components: {svd.explained_variance_ratio_.sum():.2%}")

# --- 2. Create the FULL Predicted Ratings Matrix (The Power of SVD) ---
# This is the most efficient step. We predict for ALL users and ALL movies at once.
# We add the user_means back to the dot product of the factor matrices.
# Note: user_means needs to be reshaped to align for broadcasting.
user_means_reshaped = user_means.values.reshape(-1, 1)
svd_predicted_ratings = np.dot(user_factors, item_factors) + user_means_reshaped

# Convert the numpy array back to a DataFrame with the correct index and columns
svd_preds_df = pd.DataFrame(svd_predicted_ratings, index=train_matrix.index, columns=train_matrix.columns)
print("Full SVD prediction matrix created successfully.")


# --- 3. Create an EFFICIENT Recommendation Function ---
def get_svd_recommendations(user_id, preds_df, train_matrix_df, movies_df, n=10):
    """
    Generates top N SVD-based recommendations with high efficiency by using the pre-computed predictions matrix.
    """
    if user_id not in preds_df.index:
        print(f"User {user_id} not found.")
        return pd.DataFrame()

    # Get all predicted ratings for the user from the pre-computed DataFrame
    user_predictions = preds_df.loc[user_id]

    # Get the list of movies the user has already rated from the original training matrix
    user_rated_movies = train_matrix_df.loc[user_id][train_matrix_df.loc[user_id] > 0].index

    # Filter out the movies the user has already seen and sort the rest by predicted rating
    unseen_predictions = user_predictions.drop(user_rated_movies).sort_values(ascending=False)
    
    # Get the top N recommendations
    top_n_recs = unseen_predictions.head(n)
    
    # --- HELPFUL OUTPUT ---
    rec_df = top_n_recs.reset_index()
    rec_df.columns = ['movie_id', 'predicted_rating']
    
    # Perform a single, efficient merge to get movie titles
    rec_df = rec_df.merge(movies_df, on='movie_id', how='left')
    
    return rec_df[['movie_id', 'title', 'predicted_rating']]


# --- 4. Display the Explained Variance Plots (Your code, now part of this block) ---
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle('SVD Explained Variance Analysis', fontsize=16, fontweight='bold')
# Plot 1: Explained variance per component
axes[0].bar(range(1, n_components + 1), svd.explained_variance_ratio_)
axes[0].set_title('Explained Variance per Component')
axes[0].set_xlabel('SVD Component')
axes[0].set_ylabel('Explained Variance Ratio')
# Plot 2: Cumulative explained variance
cum_var = np.cumsum(svd.explained_variance_ratio_)
axes[1].plot(range(1, n_components + 1), cum_var, marker='o', linestyle='--')
axes[1].set_title('Cumulative Explained Variance')
axes[1].set_xlabel('Number of Components')
axes[1].set_ylabel('Cumulative Variance Ratio')
axes[1].grid(True)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# =================================================================================
# FINAL, CORRECTED EVALUATION FUNCTIONS
# =================================================================================

def calculate_rmse(test_df, prediction_function, **kwargs):
    """Calculates the Root Mean Squared Error for predicted ratings on the test set."""
    preds, trues = [], []
    test_ratings_list = test_df[['user_id', 'movie_id', 'rating']][test_df['rating'] > 0].to_records()

    for _, user_id, movie_id, true_rating in test_ratings_list:
        pred_rating = prediction_function(user_id=int(user_id), movie_id=int(movie_id), **kwargs)
        preds.append(pred_rating)
        trues.append(true_rating)
        
    return np.sqrt(mean_squared_error(trues, preds)) if preds else float('inf')


def calculate_precision_recall(recommendation_function, test_matrix_df, k_values=[5, 10, 20], **kwargs):
    """
    Calculates average Precision@K and Recall@K across all test users.
    (train_matrix is now passed via **kwargs to avoid conflicts)
    """
    precisions = {k: [] for k in k_values}
    recalls = {k: [] for k in k_values}
    users_to_evaluate = test_matrix_df[test_matrix_df.sum(axis=1) > 0].index

    for user_id in users_to_evaluate:
        relevant_items = test_matrix_df.loc[user_id][test_matrix_df.loc[user_id] > 3.5].index
        if relevant_items.empty:
            continue
            
        recommended_df = recommendation_function(user_id=user_id, n=max(k_values), **kwargs)
        recommended_items = set(recommended_df['movie_id'])

        for k in k_values:
            rec_k = set(list(recommended_items)[:k])
            hits = len(rec_k.intersection(relevant_items))
            precision = hits / k if k > 0 else 0
            recall = hits / len(relevant_items) if not relevant_items.empty else 0
            precisions[k].append(precision)
            recalls[k].append(recall)

    avg_precisions = {k: np.mean(v) if v else 0 for k, v in precisions.items()}
    avg_recalls = {k: np.mean(v) if v else 0 for k, v in recalls.items()}
    
    return avg_precisions, avg_recalls


print("Final evaluation framework functions are now ready.")

# =================================================================================
# FINAL, CORRECTED EVALUATION EXECUTION
# =================================================================================

# --- Create a DataFrame to store all evaluation results ---
results = []

# --- 1. Evaluate User-Based CF ---
print("Evaluating User-Based CF...")
user_cf_rmse = calculate_rmse(test_data, predict_user_cf_rating, user_similarity_df=user_similarity_df, train_matrix=train_matrix, user_means=user_means)
user_cf_prec, user_cf_recall = calculate_precision_recall(get_user_cf_recommendations, test_matrix, train_matrix=train_matrix, user_similarity_df=user_similarity_df, movies_df=movies, user_means=user_means)
results.append({'Model': 'User-Based CF', 'RMSE': user_cf_rmse, 'Precision@10': user_cf_prec[10], 'Recall@10': user_cf_recall[10]})


# --- 2. Evaluate Item-Based CF ---
print("Evaluating Item-Based CF...")
item_cf_rmse = calculate_rmse(test_data, predict_item_cf_rating, item_similarity_df=item_similarity_df, train_matrix=train_matrix)
item_cf_prec, item_cf_recall = calculate_precision_recall(get_item_cf_recommendations, test_matrix, train_matrix=train_matrix, item_similarity_df=item_similarity_df, movies_df=movies)
results.append({'Model': 'Item-Based CF', 'RMSE': item_cf_rmse, 'Precision@10': item_cf_prec[10], 'Recall@10': item_cf_recall[10]})


# --- 3. Evaluate SVD ---
print("Evaluating SVD...")
# Corrected SVD RMSE calculation
y_true = test_matrix.values[test_matrix.values > 0]
y_pred = svd_preds_df.values[test_matrix.values > 0]
svd_rmse = np.sqrt(mean_squared_error(y_true, y_pred))
# Corrected SVD Precision/Recall call
svd_prec, svd_recall = calculate_precision_recall(get_svd_recommendations, test_matrix, preds_df=svd_preds_df, movies_df=movies, train_matrix_df=train_matrix)
results.append({'Model': 'SVD', 'RMSE': svd_rmse, 'Precision@10': svd_prec[10], 'Recall@10': svd_recall[10]})


# --- Display the final summary table ---
results_df = pd.DataFrame(results)
print("\n--- Model Performance Summary ---")
display(results_df.set_index('Model'))
```

```python
def get_hybrid_recommendations(user_id, n=10, weights={'user': 0.3, 'item': 0.3, 'svd': 0.4}):
    """
    Generates hybrid recommendations by combining the normalized scores from all three models.
    """
    # --- 1. Get Recommendations from Each Model ---
    # We call our new, efficient recommendation functions to get a candidate pool from each.
    user_recs_df = get_user_cf_recommendations(user_id, user_similarity_df, train_matrix, movies, user_means, n=20)
    item_recs_df = get_item_cf_recommendations(user_id, train_matrix, item_similarity_df, movies, n=20)
    svd_recs_df = get_svd_recommendations(user_id, svd_preds_df, train_matrix, movies, n=20)

    # --- 2. Normalize Scores and Combine ---
    # This dictionary will store the final weighted score for each movie.
    combined_scores = {}

    # Helper function to normalize and add scores
    def add_to_scores(recs_df, model_weight, scores_dict):
        if recs_df.empty:
            return

        # --- CRITICAL STEP: NORMALIZE SCORES to a 0-1 scale ---
        min_score = recs_df['predicted_rating'].min()
        max_score = recs_df['predicted_rating'].max()
        
        # Avoid division by zero if all scores are the same
        if max_score == min_score:
            recs_df['normalized_score'] = 1.0
        else:
            recs_df['normalized_score'] = (recs_df['predicted_rating'] - min_score) / (max_score - min_score)
        
        # Add the weighted, normalized score to our dictionary
        for _, row in recs_df.iterrows():
            movie_id = row['movie_id']
            scores_dict[movie_id] = scores_dict.get(movie_id, 0) + (row['normalized_score'] * model_weight)

    add_to_scores(user_recs_df, weights['user'], combined_scores)
    add_to_scores(item_recs_df, weights['item'], combined_scores)
    add_to_scores(svd_recs_df, weights['svd'], combined_scores)
    
    # --- 3. Format and Return the Final Recommendations ---
    if not combined_scores:
        return pd.DataFrame(columns=['movie_id', 'title', 'combined_score'])

    # Sort the movies by their final combined score
    sorted_recs = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)[:n]
    
    rec_ids, final_scores = zip(*sorted_recs)
    rec_df = pd.DataFrame({'movie_id': rec_ids, 'combined_score': final_scores})
    
    # Perform a single, efficient merge to get movie titles
    rec_df = rec_df.merge(movies, on='movie_id', how='left')
    
    return rec_df[['movie_id', 'title', 'combined_score']]

print("Improved hybrid recommendation function created successfully.")
```
